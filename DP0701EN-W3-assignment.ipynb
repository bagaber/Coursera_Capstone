### import BeautifulSoup & use "request" to get the data from internet

```python
from bs4 import BeautifulSoup
import requests
```

```python
wiki = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(wiki,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

#### Create a  dataframe that consists of three columns: PostalCode, Borough, and Neighborhood

```python
df = pd.read_html(tab)
data=df[0]
data.head()
```

```python
#to matach the exact column name
data.rename(columns={'Postal code':'Postalcode'},inplace=True)

data.head()
```

#### Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
data = data[data.Borough != 'Not assigned']
data.head()
```

#### Many neighborhood can exist in one postal code area. Rows to be combined into one row with the neighborhoods separated with a comma

```python
data = data.groupby(['Postalcode','Borough'], sort=False).agg(', '.join)
data.reset_index(inplace=True)
data
```

#### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
data['Neighborhood'] = np.where(data['Neighborhood'] == 'Not assigned',data['Borough'], data['Neighborhood'])
data
```

## filtering by Toronto

### using Folium to visualize the the neighborhood

```python
#Visualizing all the Neighbourhoods of the above data frame using Folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df['Latitude'],df['Longitude'],df['Borough'],df['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
k=5
toronto_clustering = df.drop(['Postalcode','Borough','Neighborhood'],1)
toronto_clustering

kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df['Latitude'], df['Longitude'], df['Neighborhood'], df['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
